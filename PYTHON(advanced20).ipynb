**Q1. Compare and contrast the float and Decimal classes' benefits and drawbacks.**

-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------


- **Float:**
  - Benefits: Efficient, widely supported, suitable for most general-purpose numerical computations.
  - Drawbacks: Limited precision, may result in rounding errors, not suitable for financial or precise decimal calculations.

- **Decimal:**
  - Benefits: Arbitrary precision, exact decimal representation, suitable for financial calculations and situations requiring high precision.
  - Drawbacks: Less efficient than floats, more memory consumption, not ideal for all general-purpose numerical computations.

-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

**Q2. Decimal('1.200') and Decimal('1.2') are two objects to consider. In what sense are these the same object? Are these just two ways of representing the exact same value, or do they correspond to different internal states?**

----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

These are two different objects. While they represent the same value mathematically (1.2), they correspond to different internal states because the number of decimal places in the representation is different.

**Q3. What happens if the equality of Decimal('1.200') and Decimal('1.2') is checked?**

----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

If you check for equality between `Decimal('1.200')` and `Decimal('1.2')`, the result will be `True`. Decimal objects are designed to consider their mathematical values when comparing for equality, not their internal representation.

**Q4. Why is it preferable to start a Decimal object with a string rather than a floating-point value?**

-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

Starting a Decimal object with a string ensures that the exact decimal representation is used. Floating-point values, when converted to Decimal directly, may result in small rounding errors due to the inherent limitations of binary floating-point representation.

**Q5. In an arithmetic phrase, how simple is it to combine Decimal objects with integers?**

-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

It is very simple to combine Decimal objects with integers in arithmetic operations. Decimal objects can be used seamlessly with integer values, and the result will be a Decimal object with the appropriate precision.

**Q6. Can Decimal objects and floating-point values be combined easily?**

-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

Decimal objects and floating-point values can be combined, but it's important to be cautious. When combining them, the result will be a Decimal object, preserving Decimal's precision. However, there may still be rounding errors if the floating-point value has limited precision.

**Q10. How do the Fraction class and the integer type (int) relate to each other? Containment or inheritance?**

-----------------------------------------------------------------------------------------------------------------------------------------------------------------------
The Fraction class and the integer type (int) do not have a direct inheritance relationship. However, the Fraction class can be used to represent integers exactly. When you create a Fraction object with an integer as the numerator and 1 as the denominator, it effectively represents an integer. For example, `Fraction(5, 1)` represents the integer 5.
