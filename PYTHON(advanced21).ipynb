**Q1. What is a probability distribution, exactly? If the values are meant to be random, how can you predict them at all?**

------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

A probability distribution is a mathematical function or description that shows the likelihood of various outcomes or values in a random experiment or event. It provides a way to model uncertainty and predict the relative frequencies of different outcomes. While individual values in a probability distribution may be random, the distribution itself is a statistical model that quantifies the probabilities of various outcomes based on the underlying random process.

------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

**Q2. Is there a distinction between true random numbers and pseudo-random numbers, if there is one? Why are the latter considered “good enough”?**

------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

Yes, there is a distinction. True random numbers are generated from a genuinely unpredictable source, such as radioactive decay or atmospheric noise, and are not influenced by any algorithm. Pseudo-random numbers, on the other hand, are generated by deterministic algorithms and appear random but are not truly random.

Pseudo-random numbers are considered "good enough" for most practical applications because they exhibit statistical properties of randomness. They are also reproducible, meaning that if you know the initial state (seed) of the random number generator, you can reproduce the sequence. This predictability and repeatability are often desirable features in computer simulations and scientific experiments.

**Q3. What are the two main factors that influence the behaviour of a "normal" probability distribution?**

------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------


The two main factors that influence the behavior of a normal probability distribution (also known as a Gaussian distribution) are:

1. **Mean (μ)**: The mean represents the center or average value of the distribution. It influences the location of the peak of the distribution.

2. **Standard Deviation (σ)**: The standard deviation measures the spread or dispersion of the values around the mean. A larger standard deviation indicates greater variability in the data, resulting in a wider distribution.

------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

**Q4. Provide a real-life example of a normal distribution.**

------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

One common real-life example of a normal distribution is the distribution of heights in a population. In many populations, human heights tend to follow a roughly normal distribution, with most people clustered around the average height, and progressively fewer individuals as you move away from the mean in either direction (taller or shorter).

**Q5. In the short term, how can you expect a probability distribution to behave? What do you think will happen as the number of trials grows?**

------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

In the short term, a probability distribution may not exhibit its expected behavior because random events can introduce variability. However, as the number of trials or observations increases, the distribution tends to converge to its expected behavior. This is a fundamental concept in statistics known as the Law of Large Numbers, which states that the average of a large number of independent and identically distributed random variables approaches the expected value of the distribution.

**Q6. What kind of object can be shuffled by using random.shuffle?**

------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

The `random.shuffle` function in Python can be used to shuffle the elements of a mutable sequence object, such as a list. Lists are commonly shuffled using this function.

**Q7. Describe the math package's general categories of functions.**

------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

The `math` package in Python provides various mathematical functions and constants. Its general categories of functions include:
- Basic mathematical operations (e.g., addition, subtraction, multiplication, division).
- Trigonometric functions (e.g., sine, cosine, tangent).
- Exponential and logarithmic functions (e.g., exponentiation, logarithms).
- Mathematical constants (e.g., pi, e).

------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

**Q8. What is the relationship between exponentiation and logarithms?**

------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

Exponentiation and logarithms are inverse operations of each other. If you have a number `x`, raising it to the power of `y` (x^y) is an exponentiation operation. The logarithm of `x` with base `y` (log_y(x)) undoes the exponentiation operation and tells you what exponent `y` you need to raise `y` to in order to obtain `x`.

**Q9. What are the three logarithmic functions that Python supports?**

Python supports the following three logarithmic functions in the `math` module:
1. `math.log(x[, base])`: Computes the natural logarithm (base e) of `x`. If the optional `base` argument is provided, it computes the logarithm of `x` with the specified base.
2. `math.log10(x)`: Computes the base-10 logarithm (common logarithm) of `x`.
3. `math.log2(x)`: Computes the base-2 logarithm of `x`.
