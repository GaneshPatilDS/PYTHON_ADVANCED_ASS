# ANS :

**Q1. Define the relationship between a class and its instances. Is it a one-to-one or a one-to-many partnership, for example?**

----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

The relationship between a class and its instances is a one-to-many partnership. A class defines a blueprint or template for creating objects (instances). You can create multiple instances of a class, each with its own unique set of attributes and behaviors, based on the class's definition.

---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

**Q2. What kind of data is held only in an instance?**

----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

Data held only in an instance is typically referred to as instance-specific data or instance variables. These variables store unique data for each instance of the class. They are defined within the methods of the class and are specific to that instance.

----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

**Q3. What kind of knowledge is stored in a class?**

----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

A class stores knowledge in the form of attributes and methods. Attributes (also called class variables) store shared data among all instances of the class. Methods define the behaviors or functions that instances of the class can perform.

----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

**Q4. What exactly is a method, and how is it different from a regular function?**

----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

A method is a function that is associated with a class. It is defined within the class and operates on the data contained within instances of the class. The key difference between a method and a regular function is that methods are bound to a class and can access and modify the class's attributes and data.

----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

**Q6. How much encapsulation (making instance or class variables private) does Python support?**


----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------


Python supports encapsulation by convention rather than strict enforcement. It uses naming conventions to indicate the level of visibility and access control:

- Variables or methods prefixed with a single underscore (e.g., `_variable`) are considered "protected" and should not be accessed directly from outside the class.
- Variables or methods prefixed with double underscores (e.g., `__variable`) are considered "private" and have name mangling to make them less accessible.

However, Python does not provide strict access control like some other languages. It relies on developers to follow naming conventions and respect encapsulation.

----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

**Q7. How do you distinguish between a class variable and an instance variable?**

----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

Class variables are shared among all instances of a class and are defined at the class level. Instance variables are specific to each instance and are defined within methods using the `self` keyword.

----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

**Q8. When, if ever, can self be included in a class's method definitions?**

----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

The `self` parameter should be included in all instance method definitions within a class. It represents the instance itself and allows you to access and modify its attributes and call other methods.

**Q9. What is the difference between the `__add__` and the `__radd__` methods?**

----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

- `__add__`: This method is called when you use the `+` operator with an object on the left side. It defines the behavior for addition when the object is on the left.
- `__radd__`: This method is called when you use the `+` operator with an object on the right side. It defines the behavior for addition when the object is on the right.

**Q10. When is it necessary to use a reflection method? When do you not need it, even though you support the operation in question?**

---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
Reflection methods (e.g., `__getattr__` and `__setattr__`) are used to customize attribute access and modification. They are necessary when you want to intercept and customize attribute-related operations. You don't need them if you don't have specific customization requirements for attribute access.


---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
**Q11. What is the `__iadd__` method called?**

----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

The `__iadd__` method is called for the `+=` operator. It defines the behavior of in-place addition when using `+=`.

----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
**Q12. Is the `__init__` method inherited by subclasses? What do you do if you need to customize its behavior within a subclass?**

The `__init__` method is not automatically inherited by subclasses, but you can call the parent class's `__init__` method within the subclass using `super().__init__()`. If you need to customize its behavior in a subclass, you can override it in the subclass by providing a new implementation while calling the parent class's `__init__` as needed.
