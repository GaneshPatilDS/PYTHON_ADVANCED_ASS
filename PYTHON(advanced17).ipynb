# ANS:


---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

**Q1. Explain the difference between greedy and non-greedy syntax with visual terms in as few words as possible. What is the bare minimum effort required to transform a greedy pattern into a non-greedy one? What characters or characters can you introduce or change?**


- **Greedy:** Matches as much as possible (e.g., `.*`).
- **Non-greedy (Lazy):** Matches as little as possible (e.g., `.*?`).
- To transform a greedy pattern into non-greedy, add `?` after the quantifier (e.g., `*` to `*?`).

--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

**Q2. When exactly does greedy versus non-greedy make a difference? What if you're looking for a non-greedy match but the only one available is greedy?**

--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

Greedy vs. non-greedy matters when you have patterns that can match multiple substrings. Non-greedy is used to capture the shortest possible match. If only a greedy match is available, you might get a longer match than desired.

---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

**Q3. In a simple match of a string, which looks only for one match and does not do any replacement, is the use of a nontagged group likely to make any practical difference?**

---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

In a simple match without replacement, a non-tagged group may not make a practical difference as you are not capturing and using the matched group explicitly.

---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

**Q4. Describe a scenario in which using a nontagged category would have a significant impact on the program's outcomes.**

--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

A scenario where non-tagged groups matter is when you're using regex for more complex operations like extracting specific data from a text document. Non-tagged groups help in capturing and processing portions of the matched text.

---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
**Q5. Unlike a normal regex pattern, a look-ahead condition does not consume the characters it examines. Describe a situation in which this could make a difference in the results of your programme.**

----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------


Look-ahead conditions are useful when you need to find a pattern but don't want to include it in the match. For example, finding email addresses in a text without including the "@" symbol in the match.

---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
**Q6. In standard expressions, what is the difference between positive look-ahead and negative look-ahead?**

---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
- **Positive Look-Ahead (`(?=...)`):** Matches if the contained pattern exists ahead in the text.
- **Negative Look-Ahead `(?!)`:** Matches if the contained pattern does not exist ahead in the text.


---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

**Q7. What is the benefit of referring to groups by name rather than by number in a standard expression?**

---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

Referring to groups by name makes your code more readable and maintainable. It's especially helpful when dealing with complex patterns with many capturing groups, as it's less error-prone and more self-documenting.

----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

**Q8. Can you identify repeated items within a target string using named groups, as in "The cow jumped over the moon"?**

----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

Named groups can be used to capture and identify repeated items in a target string. For example, you can use `(?P<word>\w+)` to capture and name words in the given string.


---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
**Q9. When parsing a string, what is at least one thing that the Scanner interface does for you that the re.findall feature does not?**

---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

The Scanner interface can be more efficient when parsing large strings as it processes the string incrementally, allowing you to take action as soon as a match is found, rather than finding all matches upfront, which `re.findall` does.

----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

**Q10. Does a scanner object have to be named scanner?**

----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

No, a scanner object doesn't have to be named "scanner." You can choose any valid variable name for your scanner object. The name you choose should be meaningful and reflect its purpose in your code for clarity.
