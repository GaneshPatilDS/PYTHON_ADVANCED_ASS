**Q1. What are the benefits of the built-in array package, if any?**

-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

The built-in `array` package in Python provides a way to create arrays, which are similar to lists but can store elements of the same data type efficiently in contiguous memory. Some benefits of the `array` package include more memory-efficient storage compared to lists, and basic array manipulation functions.


-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
**Q2. What are some of the array package's limitations?**

----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

The `array` package's limitations include its lack of extensive functionality for numerical computations and data manipulation when compared to more specialized packages like NumPy. It primarily focuses on basic array creation and manipulation.

-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

**Q3. Describe the main differences between the array and numpy packages.**

-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

- NumPy is a powerful library for numerical computations, while the `array` package is a basic built-in package for creating arrays.
- NumPy provides a wide range of mathematical and array manipulation functions, whereas the `array` package offers limited functionality.
- NumPy arrays are more versatile and can handle multi-dimensional data, while `array` is limited to one-dimensional arrays.
- NumPy is highly optimized for performance, making it suitable for scientific and data analysis tasks.

**Q4. Explain the distinctions between the empty, ones, and zeros functions.**

------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

- `numpy.empty(shape, dtype)`: Creates a new array without initializing its values. It allocates memory but does not set values, making it faster than `zeros` or `ones`. The content of the array is undefined and depends on the state of the memory.
- `numpy.zeros(shape, dtype)`: Creates a new array filled with zeros. You specify the shape (dimensions) of the array and the data type.
- `numpy.ones(shape, dtype)`: Creates a new array filled with ones. Similar to `zeros`, you specify the shape and data type.

**Q5. In the fromfunction function, which is used to construct new arrays, what is the role of the callable argument?**

------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

The `fromfunction` function in NumPy is used to construct a new array by applying a function to each element of the array. The callable argument represents the function that is applied to calculate the values of the array based on their indices. It defines the relationship between the indices and the values in the resulting array.

**Q6. What happens when a numpy array is combined with a single-value operand (a scalar, such as an int or a floating-point value) through addition, as in the expression A + n?**

------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

When a NumPy array `A` is combined with a scalar `n` through addition (`A + n`), the scalar value `n` is broadcasted to every element in the array `A`. Each element in `A` is incremented by the scalar value `n`.

------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

**Q7. Can array-to-scalar operations use combined operation-assign operators (such as += or *=)? What is the outcome?**

------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

Yes, array-to-scalar operations can use combined operation-assign operators like `+=` and `*=`. When you use `+=`, it adds the scalar value to each element in the array in place. Similarly, `*=` multiplies each element by the scalar value in place, modifying the original array.

-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

**Q8. Does a numpy array contain fixed-length strings? What happens if you allocate a longer string to one of these arrays?**

----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

Yes, a NumPy array can contain fixed-length strings. If you allocate a longer string to one of these arrays, the longer string will be truncated to fit the fixed length specified by the array's data type. No error is raised, but data beyond the specified length will be lost.

**Q9. What happens when you combine two numpy arrays using an operation like addition (+) or multiplication (*)? What are the conditions for combining two numpy arrays?**

------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

When you combine two NumPy arrays using operations like addition or multiplication, element-wise operations are performed. The arrays must have compatible shapes for these operations to work correctly. Compatible shapes either have the same dimensions or can be broadcasted to the same shape. Broadcasting is a mechanism that allows NumPy to perform element-wise operations on arrays with different shapes under certain conditions.

-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

**Q10. What is the best way to use a Boolean array to mask another array?**

-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

The best way to use a Boolean array as a mask to filter another array is by indexing. You can apply the Boolean mask directly to the target array to select specific elements based on the True/False values in the mask. For example:

```python
import numpy as np

mask = np.array([True, False, True])
data = np.array([1, 2, 3])

filtered_data = data[mask]  # Selects elements where mask is True
```

**Q11. What are three different ways to get the standard deviation of a wide collection of data using both standard Python and its packages? Sort the three of them by how quickly they execute.**

------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

Three ways to calculate the standard deviation:

1. Using NumPy: NumPy's `numpy.std(data)` function is efficient and widely used for calculating the standard deviation of data stored in a NumPy array. This method is fast and convenient.

2. Using the statistics module (Standard Library): The `statistics.stdev(data)` function can be used for calculating the standard deviation of a list of data. It's less efficient than NumPy for large datasets.

3. Manual Calculation: You can manually calculate the standard deviation by first calculating the mean and then computing the square root of the variance. This method is the slowest and less preferred for large datasets.

**Q12. What is the dimensionality of a Boolean mask-generated array?**

The dimensionality of a Boolean mask-generated array depends on the shape of the Boolean mask. If the Boolean mask is one-dimensional, the resulting array will have the same dimensionality. If the Boolean mask is multi-dimensional, the resulting array will also have the same number of dimensions. The True/False values in the mask determine which elements are included in the resulting array.
